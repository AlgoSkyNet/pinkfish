```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from talib.abstract import *
import pinkfish as pf

# Double the DPI, so we are making 2x plots:
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

Some global data

```python
#symbol = '^GSPC'
symbol = 'SPY'
capital = 10000
start = datetime.datetime(1950, 1, 1)
end = datetime.datetime.now()
```

Read spy time series data.  Use cached version if it exists, 
otherwise retrive, cache, then read.

```python
ts = pf.fetch_timeseries(symbol, use_cache=False)
ts.tail()
```

Select timeseries between start and end. Subtract a trading year from start to allow time for long term technical indicators to become valid. Back adjust prices relative to adj_close for dividends and splits (not necessary for S&P 500 index, but add it here for completeness)

```python
ts = pf.select_tradeperiod(ts, start, end, use_adj=True)
ts.head()
```

200 day MA

```python
sma200 = SMA(ts, timeperiod=200)
sma200 = np.round(sma200, decimals=2)
ts['sma200'] = sma200
```

```python
ts.tail()
```

ATR

```python
atr = ATR(ts, timeperiod=14)
atr = np.round(atr, decimals=2)
ts['atr'] = atr
```

```python
ts.tail()
```

Get documentation for ATR

```python
print(Function('atr').info)
```

5 day high and low

```python
high5 = pd.rolling_max(ts.high, 5)
low5 = pd.rolling_min(ts.low, 5)
ts['high5'] = high5
ts['low5'] = low5
```

```python
ts.tail()
```

RSI

```python
rsi2 = RSI(ts, timeperiod=2)
rsi2 = np.round(rsi2, decimals=2)
ts['rsi2'] = rsi2
c2rsi2 = pd.rolling_sum(ts.rsi2, 2)
c2rsi2 = np.round(c2rsi2, decimals=2)
ts['c2rsi2'] = c2rsi2
```

```python
print(Function('rsi').info)
```

```python
ts.tail()
```

Midpoint

```python
mp = (ts.high + ts.low)/2
mp = np.round(mp, decimals=2)
ts['mp'] = mp
```

```python
ts.tail()
```

SMA10 of midpoint

```python
sma10mp = pd.rolling_mean(ts.mp, 10)
sma10mp = np.round(sma10mp, decimals=2)
ts['sma10mp'] = sma10mp
ts.tail()
```

Standard Deviation

```python
sd = pd.rolling_std(ts.mp, 10)
ts['sd'] = sd
ts.tail()
```

```python
upper = ts.sma10mp + ts.sd*2
lower = ts.sma10mp - ts.sd*2
upper = np.round(upper, decimals=2)
lower = np.round(lower, decimals=2)
ts['upper'] = upper
ts['lower'] = lower
ts.tail()
```

```python
df = ts['2015-01-01':]
df.is_copy = False
df.head()
```

```python
import itable
pt = itable.PrettyTable(df, tstyle=itable.TableStyle(theme="theme1"), center=True, header_row=True, rpt_header=20)

pt.update_col_header_style(format_function=lambda x: x.upper(), text_align='right')
pt.update_row_header_style(format_function=lambda x: pd.to_datetime(str(x)).strftime('%Y/%m/%d'), text_align='right')

for col in range(pt.num_cols):
    if pt.df.columns[col] == 'volume':
        pt.update_cell_style(cols=[col], format_function=lambda x: format(x, '.0f'), text_align='right')
    else:
        pt.update_cell_style(cols=[col], format_function=lambda x: format(x, '.2f'), text_align='right')

for row in range(pt.num_rows):
    if row == 0:
        continue
    if (pt.df['high5'][row] == pt.df['high'][row]) and \
       (pt.df['high5'][row] > pt.df['high'][row-1]):
        col = df.columns.get_loc("high5")    
        pt.update_cell_style(rows=[row], cols=[col], color='blue')
    if (pt.df['low5'][row] == pt.df['low'][row]) and \
       (pt.df['low5'][row] < pt.df['low'][row-1]):
        col = df.columns.get_loc("low5")
        pt.update_cell_style(rows=[row], cols=[col], color='maroon')          
```

```python
def print_full(x):
    pd.set_option('display.max_rows', len(df))
    display(x)
    pd.reset_option('display.max_rows')

print_full(pt)
```
